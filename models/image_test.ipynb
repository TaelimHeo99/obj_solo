```python
# ğŸ” COCO ì˜ëª»ëœ íŒŒì¼ëª… íƒì§€ (percent-encoding, ê·œì¹™ ìœ„ë°˜)

from pathlib import Path
import re

# 1) ë°ì´í„°ì…‹ ê²½ë¡œ
root = Path("/home/taelim/workspace/obj_solo/datasets/coco")
images_dir = root / "images"
print(f"ğŸ“‚ Scan: {images_dir.resolve()}")

# 2) ìŠ¤ìº” ëŒ€ìƒ í™•ì¥ì
jpgs = (
    list(images_dir.rglob("*.jpg"))  +
    list(images_dir.rglob("*.jpeg")) +
    list(images_dir.rglob("*.JPG"))  +
    list(images_dir.rglob("*.JPEG"))
)

# 3) íŒ¨í„´ ì •ì˜
valid_re = re.compile(r'^\d{12}\.jpg$', re.IGNORECASE)  # COCO ê·œì¹™: 12ìë¦¬ ìˆ«ì + .jpg
pct_re   = re.compile(r'%[0-9A-Fa-f]{2}')               # í¼ì„¼íŠ¸ ì¸ì½”ë”©: %20, %28, %29, ...
paren_sp = re.compile(r'[()\s]')                        # ê´„í˜¸/ê³µë°± í¬í•¨ ì—¬ë¶€

percent_encoded = []
has_paren_space = []
invalid_name    = []  # ê·œì¹™ ìœ„ë°˜(12ìë¦¬ ìˆ«ì+.jpg ì•„ë‹˜)

for p in jpgs:
    name = p.name
    if pct_re.search(name):
        percent_encoded.append(str(p))
    if paren_sp.search(name):
        has_paren_space.append(str(p))
    if not valid_re.fullmatch(name):
        invalid_name.append(str(p))

# 4) ì¤‘ë³µ ì œê±°(ìˆœì„œ ë³´ì¡´)
def uniq(seq):
    seen = set(); out = []
    for x in seq:
        if x not in seen:
            seen.add(x); out.append(x)
    return out

percent_encoded = uniq(percent_encoded)
has_paren_space = uniq(has_paren_space)
invalid_name    = uniq(invalid_name)

# 5) ì¶œë ¥
def dump(title, items, limit=None):
    print(f"\n=== {title} ({len(items)}) ===")
    if not items:
        return
    if limit is None or len(items) <= limit:
        print("\n".join(items))
    else:
        print("\n".join(items[:limit]))
        print(f"... (+{len(items)-limit} more)")

dump("Percent-encoded names (e.g., %29, %28, %20)", percent_encoded, limit=2000)
dump("Names containing parentheses/spaces", has_paren_space, limit=2000)
dump("Invalid names (not 12 digits + .jpg)", invalid_name, limit=2000)

# (ì˜µì…˜) ê²°ê³¼ë¥¼ íŒŒì¼ë¡œ ì €ì¥í•˜ê³  ì‹¶ìœ¼ë©´ ì•„ë˜ ì£¼ì„ í•´ì œ
# out_dir = Path("coco_name_issues"); out_dir.mkdir(exist_ok=True)
# (out_dir / "percent_encoded.txt").write_text("\n".join(percent_encoded), encoding="utf-8")
# (out_dir / "has_paren_space.txt").write_text("\n".join(has_paren_space), encoding="utf-8")
# (out_dir / "invalid_name.txt").wri
```

```python
# ğŸ” YOLO ë¼ë²¨/ë¦¬ìŠ¤íŠ¸ ë¬´ê²°ì„± ì ê²€ (COCO/YOLO í¬ë§·)
# - ë¼ë²¨(.txt) â†” ì´ë¯¸ì§€(.jpg) ë§¤ì¹­ ê²€ì¦
# - ë¼ë²¨ íŒŒì¼ ë‚´ìš© í˜•ì‹/ë²”ìœ„ ê²€ì‚¬ (class, x y w h âˆˆ [0,1], w,h>0)
# - ë¹„ì–´ìˆëŠ” ë¼ë²¨, ì˜ëª»ëœ ë¼ë²¨ ë¼ì¸, í´ë˜ìŠ¤ ë²”ìœ„ ì´ˆê³¼, NaN/Inf ë“±
# - ì´ë¯¸ì§€/ë¼ë²¨ ê³ ì•„(orphan) íŒŒì¼
# - í¼ì„¼íŠ¸ì¸ì½”ë”©/ê³µë°±/ê´„í˜¸ í¬í•¨ íŒŒì¼ëª… ê²½ê³ 
# - (ì˜µì…˜) train/val ë¦¬ìŠ¤íŠ¸ íŒŒì¼ì˜ ê²½ë¡œ ì¡´ì¬ ì—¬ë¶€ ì ê²€

from pathlib import Path
import re
import math

# ===== ì„¤ì • =====
ROOT = Path("/home/taelim/workspace/obj_solo/datasets/coco")
IMAGES_SUBDIR = "images"      # ex) images, images/train2017 ë“± ìƒìœ„í´ë”ë¥¼ ì§€ì •í•˜ë©´ í•˜ìœ„ê¹Œì§€ rglob
LABELS_SUBDIR = "labels"      # ex) labels, labels/train2017 ë“±
LIST_FILES = [
    # (ì„ íƒ) ì´ë¯¸ì§€ ê²½ë¡œ ë¦¬ìŠ¤íŠ¸ íŒŒì¼(.txt) ì§€ì •. ê° ì¤„ì´ ì´ë¯¸ì§€ ê²½ë¡œ.
    # ROOT / "train2017.txt",
    # ROOT / "val2017.txt",
]
NUM_CLASSES = 80              # COCO ê¸°ë³¸ 80
VALID_IMG_EXTS = (".jpg", ".jpeg", ".JPG", ".JPEG")

# ===== í—¬í¼ =====
def is_percent_encoded(s: str) -> bool:
    return bool(re.search(r"%[0-9A-Fa-f]{2}", s))

def has_paren_or_space(s: str) -> bool:
    return bool(re.search(r"[()\s]", s))

def map_img_to_label(img_path: Path, images_dir: Path, labels_dir: Path) -> Path:
    # YOLO ê´€ë¡€: images/.../file.jpg -> labels/.../file.txt (ë™ì¼ ìƒëŒ€ê²½ë¡œ)
    rel = img_path.resolve().relative_to(images_dir.resolve())
    return (labels_dir / rel).with_suffix(".txt")

def map_label_to_img(lbl_path: Path, images_dir: Path, labels_dir: Path) -> Path:
    rel = lbl_path.resolve().relative_to(labels_dir.resolve())
    # ê¸°ë³¸ì€ .jpgë¡œ ë§¤í•‘ (í•„ìš”ì‹œ .jpegë„ í™•ì¸)
    return (images_dir / rel).with_suffix(".jpg")

def parse_label_line(line: str):
    # í•œ ì¤„ì— ì—¬ëŸ¬ ê°ì²´ê°€ ì˜¤ëŠ” í˜•ì‹ë„ ìˆìœ¼ë¯€ë¡œ ìŠ¤í˜ì´ìŠ¤ë¡œ split
    # YOLOëŠ” ë³´í†µ í•œ ì¤„ë‹¹ í•œ ê°ì²´ì´ì§€ë§Œ, ë‹¤ê°ì²´ í¬ë§·ë„ ëŒ€ë¹„: í† í°ìˆ˜ê°€ 5ì˜ ë°°ìˆ˜ì—¬ì•¼ í•¨.
    toks = line.strip().split()
    if len(toks) == 0:
        return []  # ë¹ˆ ì¤„
    if len(toks) % 5 != 0:
        raise ValueError(f"Token count {len(toks)} is not a multiple of 5")
    objs = []
    for i in range(0, len(toks), 5):
        cls_str, x_str, y_str, w_str, h_str = toks[i:i+5]
        # classëŠ” ì •ìˆ˜, ë‚˜ë¨¸ì§€ëŠ” ì‹¤ìˆ˜
        cls = int(cls_str)
        x = float(x_str); y = float(y_str); w = float(w_str); h = float(h_str)
        # NaN/Inf ì²´í¬
        for v in (x, y, w, h):
            if math.isnan(v) or math.isinf(v):
                raise ValueError("NaN/Inf detected")
        objs.append((cls, x, y, w, h))
    return objs

# ===== ìŠ¤ìº” ëŒ€ìƒ ìˆ˜ì§‘ =====
images_dir = (ROOT / IMAGES_SUBDIR).resolve()
labels_dir = (ROOT / LABELS_SUBDIR).resolve()
assert images_dir.exists(), f"Images dir not found: {images_dir}"
assert labels_dir.exists(), f"Labels dir not found: {labels_dir}"

imgs = [p for ext in VALID_IMG_EXTS for p in images_dir.rglob(f"*{ext}")]
lbls = list(labels_dir.rglob("*.txt"))

img_set = set(p.resolve() for p in imgs)
lbl_set = set(p.resolve() for p in lbls)

# ===== íŒŒì¼ëª… ì´ìƒì¹˜(í¼ì„¼íŠ¸ì¸ì½”ë”©/ê³µë°±/ê´„í˜¸) =====
weird_name_files = []
for p in list(img_set) + list(lbl_set):
    name = p.name
    if is_percent_encoded(name) or has_paren_or_space(name):
        weird_name_files.append(str(p))

# ===== ë§¤í•‘ ê´€ê³„ ì ê²€: ê³ ì•„ ì´ë¯¸ì§€/ë¼ë²¨ =====
orphan_images = []  # ë¼ë²¨ ì—†ëŠ” ì´ë¯¸ì§€
orphan_labels = []  # ì´ë¯¸ì§€ ì—†ëŠ” ë¼ë²¨

for ip in img_set:
    lp = map_img_to_label(ip, images_dir, labels_dir)
    if not lp.exists():
        orphan_images.append(str(ip))

for lp in lbl_set:
    ip = map_label_to_img(lp, images_dir, labels_dir)
    # .jpgê°€ ì—†ìœ¼ë©´ .jpegë„ í•œ ë²ˆ ë´ì¤€ë‹¤
    if not ip.exists():
        alt = ip.with_suffix(".jpeg")
        if not alt.exists():
            orphan_labels.append(str(lp))

# ===== ë¼ë²¨ ë‚´ìš© ì ê²€ =====
empty_labels = []             # íŒŒì¼ì€ ìˆëŠ”ë° ìœ íš¨ ê°ì²´ê°€ í•˜ë‚˜ë„ ì—†ëŠ” ê²½ìš°(ë¹ˆ íŒŒì¼ ë˜ëŠ” ìœ íš¨ ë¼ì¸ ì—†ìŒ)
malformed_labels = []         # í˜•ì‹/ìˆ˜ì¹˜ ì˜¤ë¥˜
class_out_of_range = []       # í´ë˜ìŠ¤ ë²”ìœ„ ì´ˆê³¼/ìŒìˆ˜
bbox_out_of_range = []        # x,y,w,hê°€ [0,1] ë²”ìœ„ ë°–
bbox_invalid_size = []        # w<=0 ë˜ëŠ” h<=0

for lp in lbl_set:
    try:
        text = lp.read_text(encoding="utf-8", errors="replace")
    except Exception as e:
        malformed_labels.append(f"{lp} :: READ_ERROR :: {e}")
        continue

    lines = [ln for ln in text.splitlines() if ln.strip() != ""]
    if not lines:
        empty_labels.append(str(lp))
        continue

    total_objs = 0
    any_valid = False
    for ln in lines:
        try:
            objs = parse_label_line(ln)
        except Exception as e:
            malformed_labels.append(f"{lp} :: {e} :: line='{ln}'")
            continue
        if not objs:
            continue
        any_valid = True
        for (cls, x, y, w, h) in objs:
            total_objs += 1
            # class check
            if cls < 0 or cls >= NUM_CLASSES:
                class_out_of_range.append(f"{lp} :: cls={cls}")
            # range check
            if not (0.0 <= x <= 1.0 and 0.0 <= y <= 1.0 and 0.0 <= w <= 1.0 and 0.0 <= h <= 1.0):
                bbox_out_of_range.append(f"{lp} :: {x:.3f} {y:.3f} {w:.3f} {h:.3f}")
            # size check
            if w <= 0.0 or h <= 0.0:
                bbox_invalid_size.append(f"{lp} :: {x:.3f} {y:.3f} {w:.3f} {h:.3f}")

    if not any_valid:
        empty_labels.append(str(lp))

# ===== ë¦¬ìŠ¤íŠ¸ íŒŒì¼ ì ê²€ (ì„ íƒ) =====
list_missing_paths = []
list_files_not_found = []
for lf in LIST_FILES:
    lf_abs = lf if lf.is_absolute() else (ROOT / lf)
    if not lf_abs.exists():
        list_files_not_found.append(str(lf_abs))
        continue
    for i, line in enumerate(lf_abs.read_text(encoding="utf-8", errors="replace").splitlines(), 1):
        path = line.strip()
        if not path:
            continue
        p = Path(path)
        if not p.is_absolute():
            p = (lf_abs.parent / p).resolve()
        if not p.exists():
            list_missing_paths.append(f"{lf_abs.name}:{i} -> {p}")

# ===== ê²°ê³¼ ìš”ì•½ =====
def show(title, items, limit=50):
    print(f"\n=== {title} :: {len(items)} ===")
    if len(items) == 0:
        return
    for x in items[:limit]:
        print(x)
    if len(items) > limit:
        print(f"... (+{len(items)-limit} more)")

print("âœ… Scan complete.")

show("Weird names (percent-encoded or parentheses/spaces)", weird_name_files, limit=1000)
show("Orphan images (no matching label)", orphan_images, limit=2000)
show("Orphan labels (no matching image)", orphan_labels, limit=2000)
show("Empty labels (no valid objects)", empty_labels, limit=2000)
show("Malformed labels (format errors / NaN / token count not multiple of 5)", malformed_labels, limit=2000)
show("Class out of range", class_out_of_range, limit=2000)
show("BBox out of [0,1] range", bbox_out_of_range, limit=2000)
show("BBox invalid size (w<=0 or h<=0)", bbox_invalid_size, limit=2000)

if LIST_FILES:
    show("List files not found", list_files_not_found, limit=2000)
    show("List entries pointing to missing files", list_missing_paths, limit=2000)

# (ì˜µì…˜) íŒŒì¼ë¡œ ì €ì¥í•˜ê³  ì‹¶ë‹¤ë©´ ì•„ë˜ ì£¼ì„ í•´ì œ
# out = Path("yolo_label_audit"); out.mkdir(exist_ok=True)
# (out/"weird_names.txt").write_text("\n".join(weird_name_files), encoding="utf-8")
# (out/"orphan_images.txt").write_text("\n".join(orphan_images), encoding="utf-8")
# (out/"orphan_labels.txt").write_text("\n".join(orphan_labels), encoding="utf-8")
# (out/"empty_labels.txt").write_text("\n".join(empty_labels), encoding="utf-8")
# (out/"malformed_labels.txt").write_text("\n".join(malformed_labels), encoding="utf-8")
# (out/"class_out_of_range.txt").write_text("\n".join(class_out_of_range), encoding="utf-8")
# (out/"bbox_out_of_range.txt").write_text("\n".join(bbox_out_of_range), encoding="utf-8")
# (out/"bbox_invalid_size.txt").write_text("\n".join(bbox_invalid_size), encoding="utf-8")
# if LIST_FILES:
#     (out/"list_missing_paths.txt").write_text("\n".join(list_missing_paths), encoding="utf-8")
```
